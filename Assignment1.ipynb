```python
import pandas as pd
import sys
import io

from Euler_single_step_methods import euler_explicit, euler_implicit
from Taylor_series_methods import taylor_series_method
from Runge_Kutta_explicit import *
```

```python
def print_tabular(x_y_tuple):
    x_arr, y_arr = x_y_tuple
    data = {"x_n" : x_arr, "y_n" : y_arr}
    df = pd.DataFrame(data)
    print(df)
```

```python
output_buffer = io.StringIO()
original_stdout = sys.stdout
sys.stdout = output_buffer

### Problem 1

f = lambda x, y : x + y
f_y = lambda x, y : 1
y_dd = lambda x, y : 1 + x + y
y_ddd = lambda x, y : 1 + x + y

# (i) Euler's explicit method
print("(i) Euler's explicit method")
print()
#      h = 0.2
print("h = 0.2:")
print_tabular(euler_explicit(f, 0, 1, 1, 5))
print()
#      h = 0.1
print("h = 0.1:")
print_tabular(euler_explicit(f, 0, 1, 1, 10))
print()
#      h = 0.05
print("h = 0.05:")
print_tabular(euler_explicit(f, 0, 1, 1, 20))
print('\n')

# (ii) Euler's implicit method
print("(ii) Euler's implicit method")
print()
#      h = 0.2
print("h = 0.2:")
print_tabular(euler_implicit(f, f_y, 0, 1, 1, 5))
print()
#      h = 0.1
print("h = 0.1")
print_tabular(euler_implicit(f, f_y, 0, 1, 1, 10))
print()
#      h = 0.05
print("h = 0.05")
print_tabular(euler_implicit(f, f_y, 0, 1, 1, 20))
print('\n')

# (iii) Taylor Series methods
print("(iii) Taylor Series methods")
y_derivatives_2ord = [f, y_dd]
y_derivatives_3ord = [f, y_dd, y_ddd]
print()
#      2nd Order, h = 0.2
print("2nd Order, h = 0.2")
print_tabular(taylor_series_method(y_derivatives_2ord, 0, 1, 1, 5))
print()
#      2nd Order, h = 0.1
print("2nd Order, h = 0.1")
print_tabular(taylor_series_method(y_derivatives_2ord, 0, 1, 1, 10))
print()
#      2nd Order, h = 0.05
print("2nd Order, h = 0.05")
print_tabular(taylor_series_method(y_derivatives_2ord, 0, 1, 1, 20))
print()
#      3rd Order, h = 0.2
print("3rd Order, h = 0.2")
print_tabular(taylor_series_method(y_derivatives_3ord, 0, 1, 1, 5))
print()
#      3rd Order, h = 0.1
print("3rd Order, h = 0.1")
print_tabular(taylor_series_method(y_derivatives_3ord, 0, 1, 1, 10))
print()
#      3rd Order, h = 0.05
print("3rd Order, h = 0.05")
print_tabular(taylor_series_method(y_derivatives_3ord, 0, 1, 1, 20))
print('\n')

# (iv) R-K methods
print("(iv) R-K methods")
print()
#      2nd Order(Euler-Cauchy), h = 0.2
print("2nd Order(Euler-Cauchy), h = 0.2")
print_tabular(R_K2_Euler_Cauchy(f, 0, 1, 1, 5))
print()
#      2nd Order(Euler-Cauchy), h = 0.1
print("2nd Order(Euler-Cauchy), h = 0.1")
print_tabular(R_K2_Euler_Cauchy(f, 0, 1, 1, 10))
print()
#      2nd Order(Euler-Cauchy), h = 0.05
print("2nd Order(Euler-Cauchy), h = 0.05")
print_tabular(R_K2_Euler_Cauchy(f, 0, 1, 1, 20))
print()
#      2nd Order(Improved Tangent), h = 0.2
print("2nd Order(Improved Tangent), h = 0.2")
print_tabular(R_K2_Improved_Tangent(f, 0, 1, 1, 5))
print()
#      2nd Order(Improved Tangent), h = 0.1
print("2nd Order(Improved Tangent), h = 0.1")
print_tabular(R_K2_Improved_Tangent(f, 0, 1, 1, 10))
print()
#      2nd Order(Improved Tangent), h = 0.05
print("2nd Order(Improved Tangent), h = 0.05")
print_tabular(R_K2_Improved_Tangent(f, 0, 1, 1, 20))
print()
#      3rd Order(Nystrom), h = 0.2
print("3rd Order(Nystrom), h = 0.2")
print_tabular(R_K3_Nystrom(f, 0, 1, 1, 5))
print()
#      3rd Order(Nystrom), h = 0.1
print("3rd Order(Nystrom), h = 0.1")
print_tabular(R_K3_Nystrom(f, 0, 1, 1, 10))
print()
#      3rd Order(Nystrom), h = 0.05
print("3rd Order(Nystrom), h = 0.05")
print_tabular(R_K3_Nystrom(f, 0, 1, 1, 20))
print()
#      4th Order(Classical), h = 0.2
print("4th Order(Classical), h = 0.2")
print_tabular(R_K4_Classical(f, 0, 1, 1, 5))
print()
#      4th Order(Classical), h = 0.1
print("4th Order(Classical), h = 0.1")
print_tabular(R_K4_Classical(f, 0, 1, 1, 10))
print()
#      4th Order(Classical), h = 0.05
print("4th Order(Classical), h = 0.05")
print_tabular(R_K4_Classical(f, 0, 1, 1, 20))

sys.stdout = original_stdout
captured_output = output_buffer.getvalue()
output_buffer.close()
print(captured_output)
with open("answers/assignment1/Problem1.txt", "w") as file:
    file.write(captured_output)
```
